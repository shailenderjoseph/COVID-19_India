```python
import pandas as pd
import numpy as np
import math
from  geopy.geocoders import Nominatim
```

__Districwise-2020-04-03__

```python
df0403 = pd.read_excel('Districwise-2020-04-03.xlsx', header = 0)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df0403)
```

```python
df0403['District'][26] = 'Central Delhi'
df0403['District'][29] = 'North East Delhi'
df0403['District'][30] = 'North West Delhi'
df0403['District'][32] = 'South West Delhi'
df0403['District'][40] = 'Gir Somnath'
df0403['District'][74] = 'Kalburgi'
df0403['District'][77] = 'Tumkur'
df0403['District'][79] = 'Uttara Kannada'
df0403['District'][91] = 'Thiruvananthapuram'
df0403['District'][108] = 'Mumbai Suburban'
df0403['District'][110] = 'Jalgaon'
df0403['District'][150] = 'Coimbatore'
df0403['District'][153] = 'Kanchipuram'
df0403['District'][165] = 'Tirunelveli'
df0403['District'][189] = 'Bhadradri Kothagudem'
df0403['District'][187] = 'Moradabad'
df0403['District'][184] = 'Pilibhit'
df0403['District'][191] = 'Medchal'
df0403['District'][205] = 'North Pargana'
df0403['District'][206] = 'South Pargana'
#df0403.drop([148], inplace = True)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df0403)
```

```python
for i in df0403.index:
    if math.isnan(df0403['Lat'][i]):
        geolocator = Nominatim()
        city = df0403['District'][i]
        country ="India"
        loc = geolocator.geocode(city+','+ country)
        df0403['Lat'][i] = loc.latitude
        df0403['Long'][i] = loc.longitude
        print(i, "latitude is :-" ,loc.latitude,"\nlongtitude is:-" ,loc.longitude)
```

```python
df0403.to_csv("Districwise-2020-04-03.csv")
```

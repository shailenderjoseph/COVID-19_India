```python
union = pd.concat([df0327,df0329,df0330,df0401,df0402,df0403,df0404,df0405,df0406,df0408,df0409], ignore_index=True)
union
```

```python
union['Radius'] = ''
union['Color'] = ''
for e in range(0,len(union)):
    if 0<union['No of Positive cases'][e]<5:
        union['Radius'][e] = 3
        union['Color'][e] = '#56e8ab'
    if 5<=union['No of Positive cases'][e]<10:
        union['Radius'][e] = 5
        union['Color'][e] = '#428bca'
    if 10<=union['No of Positive cases'][e]<20:
        union['Radius'][e] = 7
        union['Color'][e] = '#0043fa'
    if 20<=union['No of Positive cases'][e]<50:
        union['Radius'][e] = 9
        union['Color'][e] = '#ff9d00'
    if 50<=union['No of Positive cases'][e]<100:
        union['Radius'][e] = 10
        union['Color'][e] = '#fa5c00'
    if 100<=union['No of Positive cases'][e]:
        union['Radius'][e] = 12
        union['Color'][e] = '#ff0303'
```

```python
def create_geojson_features(df):
    features = []
    
    for _, row in df.iterrows():
        feature = {
            'type': 'Feature',
            'geometry': {
                'type':'Point', 
                'coordinates':[row['Longitude'],row['Latitude']]
            },
            'properties': {
                'time': row['Date'],
                'popup': {(row['Date'],row['District']): row['No of Positive cases']}.__str__(),
                #'tooltip': tooltip,
                'style': {'color' : ''},
                'icon': 'circle',
                'iconstyle':{
                    'fillColor': row['Color'],
                    'fillOpacity': 0.8,
                    'stroke': 'true',
                    'radius': row['Radius']
                }
            }
        }
        features.append(feature)
    return features
start_geojson = create_geojson_features(union) 
start_geojson[0]
```

```python
m = folium.Map(location=[20.5937, 78.9629], zoom_start=4)
map = TimestampedGeoJson(start_geojson,
                  period = 'P1D',
                  duration = None,
                  transition_time = 1000,
                  min_speed = 0.001,
                  max_speed = 2,
                  auto_play = True).add_to(m)
m
```
